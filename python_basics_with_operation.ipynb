```python
# Basic variables and math
a = 76
b = 78
c = 89
d = 76

print(a + b)      # addition
print(c + a - d)  # mixed operation
print(d * a)      # multiplication
```

```python
# Variable naming examples
bd_age = 76
ind_age = 98   # underscore (_) allowed
ame_age = 56   # cannot start with number

print(bd_age + ame_age)
```

```python
# Different data types
name = "Tanver"
address = "kushtia"
postal_code = 7000
cost = 50.8
Good_Man = True
```

```python
# Check data types
print(type(name))
print(type(address))
print(type(postal_code))
print(type(cost))
print(type(Good_Man))
```

```python
# String searching
name = "Tanver"

print("e" in name)     # check if 'e' exists
print(name.find("e"))  # position of 'e'
print(name.index("r")) # position of 'r'
print(name.index("a")) # position of 'a'
```

```python
# Access characters by index
print(name[0])
print(name[1])
print(name[-4])
print(name[-3])
```

```python
print(address[:])        # full string
print(address[::-1])     # reverse
print(address[::1])      # normal
print(address[::2])      # skip 1 character
print(address[::3])      # skip 2 characters
```

```python
print(address[:5])     # first 5 characters
print(address[3:])     # from index 3 to end
print(address[1:5])    # index 1 to 4
print(address[0:7])    # full range
```

```python
# Reverse with step
print(address[-1:-8:-2])
```

```python
# Range example
a = range(500)
print(a)   # shows range object (0 to 499)
```

```python
# Power and remainder
print(2**4)   # 2^4 = 16
print(28 % 3) # remainder = 1
print(27 % 3) # remainder = 0
```

```python
# Floor division
print(15 // 2)  # gives 7 (no decimal)
```

```python
# Round function
print(round(3.1416))
print(round(3.50))
print(round(3.1416, 2))
```
